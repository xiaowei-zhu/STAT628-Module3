```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , confusion_matrix 
from sklearn.model_selection import train_test_split
```

```python
# load data
df = pd.read_csv("att_df.csv",index_col=0)
business = pd.read_csv('bars_business.csv')
df = pd.concat([df,pd.DataFrame(business.stars)], axis=1)
```

```python
# df.NoiseLevel = df.NoiseLevel.fillna('NoValue')
df.WiFi = df.WiFi.fillna('no')
# df.Alcohol = df.Alcohol.fillna('NoValue')
# df.replace(to_replace=None, value=np.nan, inplace=True)

df.BikeParking = df.BikeParking.fillna('False')
df.BusinessAcceptsCreditCards = df.BusinessAcceptsCreditCards.fillna('False')
df.RestaurantsReservations = df.RestaurantsReservations.fillna('False')
df.RestaurantsTakeOut = df.RestaurantsTakeOut.fillna('False')
df.RestaurantsDelivery = df.RestaurantsDelivery.fillna('False')
df.HasTV = df.HasTV.fillna('False')
df.RestaurantsGoodForGroups = df.RestaurantsGoodForGroups.fillna('False')
df.HappyHour = df.HappyHour.fillna('False')
df.OutdoorSeating = df.OutdoorSeating.fillna('False')

```

```python

# solve 'u' and transfer string 
for i in range(df.shape[0]):
  if isinstance(df.NoiseLevel[i], str) and df.NoiseLevel[i][0] == 'u':
    df.NoiseLevel[i] = df.NoiseLevel[i][1:]
  if isinstance(df.WiFi[i], str) and df.WiFi[i][0] == 'u':
    df.WiFi[i] = df.WiFi[i][1:]
  if isinstance(df.Alcohol[i], str) and df.Alcohol[i][0] == 'u':
    df.Alcohol[i] = df.Alcohol[i][1:]
  if(isinstance(df.BusinessParking[i], str)):
    df.BusinessParking[i] = eval(df.BusinessParking[i])
  if(isinstance(df.Music[i], str)):
    df.Music[i] = eval(df.Music[i])
  if(isinstance(df.Ambience[i], str)):
    df.Ambience[i] =eval(df.Ambience[i])
  # df.BikeParking[i] = eval(df.BikeParking[i])
  # df.BusinessAcceptsCreditCards[i] = eval(df.BusinessAcceptsCreditCards[i])
  df.RestaurantsReservations[i] = eval(df.RestaurantsReservations[i])
  if(isinstance(df.RestaurantsTakeOut[i], str)):
    df.RestaurantsTakeOut[i] = eval(df.RestaurantsTakeOut[i])
  if(isinstance(df.RestaurantsDelivery[i], str)):
    df.RestaurantsDelivery[i] = eval(df.RestaurantsDelivery[i])
  if(isinstance(df.HasTV[i], str)):
    df.HasTV[i] = eval(df.HasTV[i])
  # df.RestaurantsPriceRange2[i] = eval(df.RestaurantsPriceRange2[i])
  # df.RestaurantsGoodForGroups[i] = eval(df.RestaurantsGoodForGroups[i])
  df.OutdoorSeating[i] = eval(df.OutdoorSeating[i])
  # df.HappyHour[i] = eval(df.HappyHour[i])
  if(isinstance(df.NoiseLevel[i], str)):
    df.NoiseLevel[i] = eval(df.NoiseLevel[i])
  if(isinstance(df.WiFi[i], str) and df.WiFi[i]!='no'):
    df.WiFi[i] = eval(df.WiFi[i])
  if(isinstance(df.Alcohol[i], str)):
    df.Alcohol[i] = eval(df.Alcohol[i])
# # drop NaN
# df = df.dropna(axis=0, how='any')
# df = df.reset_index(drop=True)
```

```python
df.BikeParking = df.BikeParking.fillna(False)
df.BusinessAcceptsCreditCards = df.BusinessAcceptsCreditCards.fillna(False)
df.RestaurantsReservations = df.RestaurantsReservations.fillna(False)
df.RestaurantsTakeOut = df.RestaurantsTakeOut.fillna(False)
df.RestaurantsDelivery = df.RestaurantsDelivery.fillna(False)
df.HasTV = df.HasTV.fillna(False)
df.RestaurantsGoodForGroups = df.RestaurantsGoodForGroups.fillna(False)
df.HappyHour = df.HappyHour.fillna(False)
df.OutdoorSeating = df.OutdoorSeating.fillna(False)
```
